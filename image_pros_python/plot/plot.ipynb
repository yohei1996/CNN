```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['font.size']=14
import numpy as np
np.random.seed(123)
N = 100
x = np.random.rand(N)
y = 2*x + np.random.randn(N)
c= np.random.randn(N)
fig =  plt.figure(figsize= (10,8))
plt.subplot(321)
plt.scatter(x,y,c=c,cmap=plt.cm.Accent)
plt.subplot(322)
plt.scatter(x,y,c=c,cmap=plt.cm.Accent)
plt.subplot(323)
plt.scatter(x,y,c=c,cmap=plt.cm.Accent)
plt.subplot(111)
plt.scatter(x,y,c=c,cmap=plt.cm.Accent)
plt.subplot(111)
plt.show()
```
